```python
import pandas as pd

pd.set_option('display.max_columns',150)
pd.set_option('display.max_rows',150)
```

```python
df1 = pd.read_csv('hyderabad_second_type_000.csv')
df2 = pd.read_csv('hyderabad_second_type_100.csv')
df3 = pd.read_csv('hyderabad_second_type_200.csv')
df4 = pd.read_csv('hyderabad_second_type_300.csv')
df5 = pd.read_csv('hyderabad_second_type_400.csv')
df6 = pd.read_csv('hyderabad_second_type_500.csv')
```

```python
df1.head()
```

```python
frames1 = [df1,df2]
df_cat1 = pd.concat(frames1)

frames2 = [df_cat1,df3]
df_cat2 = pd.concat(frames2)

frames3 = [df_cat2,df4]
df_cat3 = pd.concat(frames3)

frames4 = [df_cat3,df5]
df_cat4 = pd.concat(frames4)

frames5 = [df_cat4,df6]
df_cat5 = pd.concat(frames5)
```
