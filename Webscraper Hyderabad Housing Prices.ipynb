```python
import re
from IPython.core.display import clear_output
from warnings import warn

import time
from time import sleep
from random import randint

import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen,Request
import csv
```

```python
headers = ({'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_5_2) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/14.1.2 Safari/605.1.15'})
```

```python
url = 'https://www.magicbricks.com/property-for-sale/residential-real-estate?&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment,Residential-House,Villa&cityName=Hyderabad'
```

```python
flats = []
boxes = []
p = 28
q = 2

for i in range(0,100):
    
    request = Request(url, headers=headers)
    response = urlopen(request)
    
    sleep(randint(8,15))
    
    html = response.read()
    html_soup = BeautifulSoup(html ,'html.parser')
    
    house_container = html_soup.find_all('span', class_ = 'domcache js-domcache-srpgtm')
    house_container1 = html_soup.find_all('div', class_="m-srp-card__summary js-collapse__content")
    house_container2 = html_soup.find_all('span', class_="hidden")

    
    for i in range(len(house_container)):
           
        try:
            price = house_container[i]['data-priced']
        except:
            price = ''
            
        try:
            isCarpet = house_container[i]['data-ca']
        except:
            isCarpet = ''    

        try:
            bedrooms = house_container[i]['data-bedroom']
        except:
            bedrooms = ''    

        try:
            builder = house_container[i]['data-companyname']
        except:
            builder = '' 

        try:
            isPaidUser = house_container[i]['data-paiduser']
        except:
            isPaidUser = ''    

        try:
            locality = house_container[i]['data-objlmtdname']
        except:
            locality = ''  

        try:
            prop_type = house_container[i]['data-objproptypeid']
        except:
            prop_type = ''  

        try:
            user_type = house_container[i]['data-susertype']
        except:
            user_type = ''  

        try:
            is_prime = house_container[i]['data-primelocprop']
        except:
            is_prime = '' 

        try:
            agent = house_container[i]['data-soname']
        except:
            agent = ''  

        try:
            link = house_container[i]['data-detailurl']
        except:
            link='' 
            
        try:
            prop_id = house_container[i]['id']
        except:
            prop_id =''    

        try:
            other = house_container1[i].text.replace('\n','*')
        except:
            other=''   
            
            
            
        
        flat = (price,isCarpet,bedrooms,builder,isPaidUser,locality,prop_type,user_type,is_prime,agent,link,
                prop_id,other)

        flats.append(flat)
        
        #end of first loop********************************************************
        
    
    for i in range(len(house_container2)):
        
        try:
            bathrooms = house_container2[i]['data-bathroom']
        except:
            bathrooms =''
            
        try:
            bedrooms2 = house_container2[i]['data-bedroom']
        except:
            bedrooms2 =''  
            
        try:
            date_created = house_container2[i]['data-createdate']
        except:
            date_created ='' 
            
        try:
            builder2 = house_container2[i]['data-devname']
        except:
            builder2 =''    
    
        try:
            floor = house_container2[i]['data-floorno']
        except:
            floor =''
        
        try:
            furnished_status = house_container2[i]['data-furnshingstatus']
        except:
            furnished_status =''  
        
        try:
            price2 = house_container2[i]['data-price']
        except:
            price2 =''   
    
        try:
            project_name = house_container2[i]['data-projectname']
        except:
            project_name =''
        
        try:
            prop_type2 = house_container2[i]['data-propertyval']
        except:
            prop_type2 ='' 
        
        try:
            isVerified = house_container2[i]['data-isverified']
        except:
            isVerified =''
        
        try:
            brand_new = house_container2[i]['data-transactiontype']
        except:
            brand_new =''
            
        try:
            user_type = house_container2[i]['data-usertype']
        except:
            user_type =''  
        
        try:
            prop_id2 = house_container2[i]['id']
        except:
            prop_id2 =''
            
    
        
        box = (bathrooms,bedrooms2,date_created,builder2,floor,furnished_status,price2,project_name,prop_type2,
               isVerified,brand_new,user_type,prop_id2)

        boxes.append(box)
      
    url = 'https://www.magicbricks.com/mbsearch/propertySearch.html?propertyType_new=10002_10003_10021_10022,10001_10017&postedSince=-1&city=2060&searchType=1&propertyType=10002,10003,10021,10022,10001,10017&disWeb=Y&pType=10002,10003,10021,10022,10001,10017&category=S&cusImgCount=0&groupstart={}&offset=0&maxOffset=47&attractiveIds=&page={}&ltrIds=54364457,54356527&preCompiledProp=&excludePropIds=&addpropertyDataSet='.format(p,q)
    # Notice the placeholders {} in the above url
    p =  p+30
    q = q+1         
```

```python
with open('hyderabad_first_type_000.csv', 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(['Price','Carpet Area','Bedrooms','Builder','Paid User','Locality','Property Type','User Type',
                    'Prime Location','Agent','Link','Prop ID','Other'])
        w.writerows(flats)
```

```python
with open('hyderabad_second_type_000.csv', 'w', newline='', encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow(['Bathrooms','Bedrooms2','Date Created','Builder2','Floor','Furnished Status','Price2',
                   'Project Name','Property Type','Verified','Brand New','User Type','Prop ID2'])
        w.writerows(boxes)
```
