```python
import pandas as pd

pd.set_option('display.max_columns',150)
pd.set_option('display.max_rows',150)
```

```python
df1 = pd.read_csv('concat_first.csv')
df2 = pd.read_csv('concat_second.csv')
```

```python
df1['Prop ID'].nunique() 
```

```python
df1.drop_duplicates(subset='Prop ID',inplace=True)
```

```python
df1.shape
```

```python
df2['Prop ID'].nunique()
```

```python
df2.drop_duplicates(subset='Prop ID',inplace=True)
```

```python
df2.shape
```

```python
df_merge = pd.merge(left=df1, right=df2, how='inner',on='Prop ID')
```

```python
df_merge.shape
```

```python
df_merge.head()
```

```python
df_merge.columns = df_merge.columns.str.replace(' ','_')
df_merge.columns = [x.lower() for x in df_merge.columns]
```

```python
df_merge.to_csv('merge_first_second_raw_10907x23.csv',index=False)
```
